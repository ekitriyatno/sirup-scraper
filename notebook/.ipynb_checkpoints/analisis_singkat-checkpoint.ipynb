```python
## Ambil data yang sudah di scrap di database kemudian kita buat ke dataframe
```

```python
import pandas as pd
import mysql.connector
import matplotlib.pyplot as plt
from rapidfuzz import process, fuzz
from datetime import time, datetime
import os

# Koneksi ke database
try:
    cnx = mysql.connector.connect(user='root', password='',
                                 host='localhost',
                                 database='rise_up',
                                 use_pure=False)
    cursor = cnx.cursor()
except mysql.connector.Error as e:
    print(f"Gagal Koneksi {e}")

cursor.execute("select * from scraped_sirup")
datasirup = cursor.fetchall()
df = pd.DataFrame(datasirup)

df.rename(columns={0:'kode_rup',1:'nama_paket',2:'nama_klpd',3:'satuan_kerja',4:'tahun_anggaran',5:'provinsi',6:'kabupaten_kota',7:'detail_lokasi',8:'volume_pekerjaan',9:'total_pagu',10:'metode_pemilihan',11:'tanggal_umumkan_paket',12:'jenis_pengadaan',13:'sumber_dana',14:'aspek_ekonomi',15:'aspek_sosial',16:'aspek_lingkungan',17:'pemanfaatan_barang_jasa_mulai',18:'pemanfaatan_barang_jasa_akhir',19:'jadwal_pelaksanaan_kontrak_mulai',20:'jadwal_pelaksanaan_kontrak_akhir',21:'jadwal_pemilihan_penyedia_mulai',22:'jadwal_pemilihan_penyedia_akhir'}, inplace=True)

# Tampilkan 5 baris pertama
df.head()
```

```python
## Membersihkan data
## Kalimat-kalimat yang membuat data kurang relevan dengan produk yang ingin kita tawarkan karena dsini fokus untuk penyediaan produk nutrisi
```

```python
#Keyword yang kurang sesuai atau kurang relevan
kata_kunci = ['rapat','meeting', 'kegiatan', 'alat','jasa','tamu','jamuan','lapangan', 'suku cadang','uang']

# Filter baris yang TIDAK mengandung kata-kata tersebut
mask = ~df['nama_paket'].str.lower().str.contains('|'.join(kata_kunci))

# Simpan hasil bersih
df_sirup = df[mask].copy()

df_sirup.head(20)
```

```python
## Mapping ke provinsi dan kota/kabupaten
```

```python
#Cek jika ada provinsi yang tidak sesuai based on data
dirfile = 'D:/SHP/Dev/Python/Script/daftar_provinsi.xlsx'
df_prov = pd.read_excel(dirfile, engine='openpyxl')

df_prov['kabupaten_kota'] = df_prov['kabupaten_kota'].str.strip().str.lower()
df_sirup['kabupaten_kota'] = df_sirup['kabupaten_kota'].str.strip().str.lower()

# Buat fungsi untuk cari nama kabupaten paling mirip
def cari_kabupaten_terdekat(kabupaten):
    hasil = process.extractOne(
        query=kabupaten,
        choices=df_prov['kabupaten_kota'],
        scorer=fuzz.token_sort_ratio
    )
    return hasil[0] if hasil else None

# Terapkan ke kolom baru
df_sirup['kabupaten_kota_match'] = df_sirup['kabupaten_kota'].apply(cari_kabupaten_terdekat)

# Lakukan join fuzzy
df_final = df_sirup.merge(df_prov, left_on='kabupaten_kota_match', right_on='kabupaten_kota', how='left', suffixes=('', '_valid'))

na = df_final[df_final['kabupaten_kota_match'].isna()]
na.head()
```

```python
plt.bar(df_final['provinsi_valid'], df_final['total_pagu'])
plt.title("Nilai Tender per Provinsi")
plt.xlabel("Provinsi")
plt.ylabel("Total Pagu")
plt.show()
```

```python
## Melihat ada berapa potensial sales yang aktif (berdasarkan tanggal pemilihan penyedia yang belum lewat) di setiap provinsi
```

```python
# Pastikan kolom tanggal berbentuk datetime
df_final['jadwal_pemilihan_penyedia_akhir'] = pd.to_datetime(df_final['jadwal_pemilihan_penyedia_akhir'], format='mixed', errors='coerce')

# Ambil data aktif (belum lewat)
hari_ini = pd.to_datetime(datetime.today().date())
df_aktif = df_final[df_final['jadwal_pemilihan_penyedia_akhir'] >= hari_ini]

#Seleksi datanya
hasil = df_aktif.groupby('provinsi_valid').agg({'nama_paket':'size','total_pagu':'sum'}).reset_index()

# Export data ke csv Simpan file ke folder output
output_dir = 'D:/SHP/Dev/ScrapySirup/scrapsirup/output'
hasil.to_csv(os.path.join(output_dir, 'result.csv'), index=False)
```
